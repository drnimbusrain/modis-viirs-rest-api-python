# Working with `xarray`

`xarray` is an open source project and Python package that aims to bring the labeled data power of pandas to the physical sciences, by providing N-dimensional variants of the core pandas data structures.

[**http://xarray.pydata.org/**](http://xarray.pydata.org/)

In this tutorial we will submit a request for a Leaf Area Index (LAI) time series from the combined MODIS Terra+Aqua product MCD43A3H for a 100x100-km area in east Tennessee. Then we'll learn how to create an `xarray.DataArray` object from the returned data, plot it, and save it as netCDF.

**Imports:**

```python
import os,json,requests
import xarray as xr
import numpy as np
import datetime as dt
%matplotlib inline
```

We will request a 3-year period from 2015-2017.

**Set subset parameters:**

```python
url = "https://modis.ornl.gov/rst/api/v1/"
header = {'Accept': 'application/json'}

prod = "MCD15A3H"
band = "Lai_500m"
lat, lng = 35, -83
sd = dt.datetime.strptime('2015-01-01', "%Y-%m-%d").date()
ed = dt.datetime.strptime('2018-01-01', "%Y-%m-%d").date()
ab, lr = 50, 50
```

**Get a list of dates using the dates function and parse to list of modis dates (i.e. AYYYYDOY):**

```python
datesurl = "".join([ url, prod, "/dates?", "latitude=", str(lat), "&longitude=", str(lng) ])
response = requests.get(datesurl, headers=header)
dates = json.loads(response.text)['dates']
print(dates[0:5])

modis_dates = [ d['modis_date'] for d in dates if all([ dt.datetime.strptime(d['calendar_date'], "%Y-%m-%d").date() >= sd,  dt.datetime.strptime(d['calendar_date'], "%Y-%m-%d").date() < ed ]) ]
print(modis_dates[0:5])
```

**Define functions to ... **
* divide `modis_dates` list into increments of 10 
* assemble url string from subset request parameters

```python
"""yield successive n-sized chunks from list l"""
def chunk(l, n):
    for i in range(0, len(l), n):
        yield l[i:i + n]

"""assemble request URL string"""
def getSubsetURL( prod , lat , lon , band, sd , ed , ab , lr ): return( "".join([
    url, prod, "/subset?",
    "latitude=", str(lat),
    "&longitude=", str(lon),
    "&band=", str(band),
    "&startDate=", str(sd),
    "&endDate=", str(ed),
    "&kmAboveBelow=", str(ab),
    "&kmLeftRight=", str(lr) ]) )
```

**Iterate over groups of dates, request subsets from the REST API, and append to a list of responses:**

```python
chunks = list(chunk(modis_dates, 10))
subsets = []
for i, c in enumerate(chunks):     
    print("[ "+str(i+1)+" / "+str(len(chunks))+" ] "+c[0]+" - "+c[-1])
    _url = getSubsetURL(prod, lat, lng, band, c[0], c[-1], ab, lr)
    _response = requests.get(_url , headers = header)
    subsets.append( json.loads(_response.text) )
```

**Use dictionary comprehension to get some spatial metadata from the first subset in our list:**

```python
meta = { key:value for key,value in subsets[0].items() if key != "subset" }
meta
```

**Iterate over the list of subsets and collect the data in a dictionary:**

```python
data = { 'dates': [], 'arrays': [] }
for i in subsets:
    for j in i['subset']:
        data['dates'].append( j['calendar_date'] )
        data['arrays'].append( np.array( j['data']).reshape(meta['nrows'], meta['ncols'] ) )
```

**Use the metadata to make lists of x and y coordinates:**          
`f(ncols,nrows): n * cellsize + corner_coordinate`

```python
dtdates = [ dt.datetime.strptime(d,"%Y-%m-%d") for d in data['dates'] ]
xcoordinates = [float(meta['xllcorner'])]+[ i*meta['cellsize']+float(meta['xllcorner']) for i in range(1,meta['ncols']) ]
ycoordinates = [float(meta['yllcorner'])]+[ i*meta['cellsize']+float(meta['yllcorner']) for i in range(1,meta['nrows']) ]
print(str([xcoordinates[0],ycoordinates[0]]))
```

**Make an `xarray.DataArray` object:**

```python
_xrDataArray = xr.DataArray( 
    name = meta['band'],
    data = np.flipud( np.dstack(data['arrays']) ),
    coords = [ np.array(ycoordinates), np.array(xcoordinates), dtdates  ],
    dims = [ "y", "x", "time" ],
    attrs = dict( units = meta['units'] )
)
_xrDataArray
```

**Remove values outside of the valid range (0 to 100) from the array and apply the scale factor:**

```python
xrDataArray = _xrDataArray.where( _xrDataArray<=100 ) * float(meta['scale'])
xrDataArray
```

**Now plotting the data is super easy:**

```python
mean = xrDataArray.mean(['x','y'])
std = xrDataArray.std(['x','y'])

mean.plot(size=6,aspect=3)
(mean-std).plot(color="gray",ls=":")
(mean+std).plot(color="gray",ls=":")
```

**Add some new coordinates to the array for indexing:**

```python
xrDataArray.coords['month'] = xrDataArray.time.dt.month
xrDataArray.coords['year'] = xrDataArray.time.dt.year
```

**Plot again!**       
Note: `xarray` doesn't allow grouping by \[`.groupby()`\] multiple dimensions so you have to first group by the year coordinate, then apply a user function \[`.apply()`\] across each year that groups by the month coordinate.

Try it by plotting the gridded mean for July of each year:

```python
monthlymean1 = xrDataArray.groupby('year').apply(lambda y: y.groupby('month').mean('time',skipna=True))
monthlymean1.isel(month=slice(6,7)).plot.pcolormesh('x','y',col='year',row='month', size=6)
```

**We just calculated the mean over the `time` dimension indexed by month and year. Now plot over the `x` and `y` dimensions:**

```python
monthlymean2 = xrDataArray.groupby('year').apply(lambda y: y.groupby('month').mean(skipna=True))
monthlymean2.sel(year=np.unique(xrDataArray.year)).plot.line(x='month',size=6,aspect=3)
```

**You can aggregate over any coordinate that you add to the xarray object. Plot the day of the year time series by year:**

```python
xrDataArray.coords['day'] = xrDataArray.time.dt.dayofyear

dailymean = xrDataArray.groupby('year').apply(lambda y: y.groupby('day').mean(skipna=True))
dailymean.sel(year=np.unique(xrDataArray.year)).plot.line(x='day',size=6,aspect=3)
```

**Finally, save as netCDF:**

```python
xrDataArray.to_netcdf('output.nc',unlimited_dims="time")
```
