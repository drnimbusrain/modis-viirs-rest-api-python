# Working with `gdal`

GDAL is a translator library for raster and vector geospatial data formats that is released under an X/MIT style Open Source license by the Open Source Geospatial Foundation. The bindings to the GDAL library are popular tools for manipulating geospatial data in Python and the foundation for most higher-level raster data manipulation packages.

[**https://www.gdal.org/**](https://www.gdal.org/)

In this tutorial we will submit a request for a Leaf Area Index (LAI) time series from the combined MODIS Terra+Aqua product MCD15A3H for a 100x100-km area in east Tennessee. Then we'll learn how to create an `gdal.Raster` object from the returned data, plot it, and save it as a GeoTIFF.

**Imports:**

````python
import os,json,requests
import numpy as np
import datetime as dt
from osgeo import gdal,osr
from matplotlib import pyplot as plt
%matplotlib inline

proj = "+proj=sinu +lon_0=0 +x_0=0 +y_0=0 +a=6371007.181 +b=6371007.181 +units=m +no_defs"
````

We will request a 3-year period from 2015-2017.

**Set subset parameters:**

````python
url = "https://modis.ornl.gov/rst/api/v1/"
header = {'Accept': 'application/json'}

prod = "MCD15A3H"
band = "Lai_500m"
lat, lng = 35, -83
sd = dt.datetime.strptime('2015-01-01', "%Y-%m-%d").date()
ed = dt.datetime.strptime('2018-01-01', "%Y-%m-%d").date()
ab, lr = 50, 50
````

**Get a list of dates using the dates function and parse to list of modis dates (i.e. AYYYYDOY):**

````python
datesurl = "".join([ url, prod, "/dates?", "latitude=", str(lat), "&longitude=", str(lng) ])
response = requests.get(datesurl, headers=header)
dates = json.loads(response.text)['dates']
print(dates[0:5])

modis_dates = [ d['modis_date'] for d in dates if all([ dt.datetime.strptime(d['calendar_date'], "%Y-%m-%d").date() >= sd,  dt.datetime.strptime(d['calendar_date'], "%Y-%m-%d").date() < ed ]) ]
print(modis_dates[0:5])
````

**Define functions to ... **
* divide `modis_dates` list into increments of 10 
* assemble url string from subset request parameters

````python
"""yield successive n-sized chunks from list l"""
def chunk(l, n):
    for i in range(0, len(l), n):
        yield l[i:i + n]

"""assemble request URL string"""
def getSubsetURL( prod , lat , lon , band, sd , ed , ab , lr ): return( "".join([
    url, prod, "/subset?",
    "latitude=", str(lat),
    "&longitude=", str(lon),
    "&band=", str(band),
    "&startDate=", str(sd),
    "&endDate=", str(ed),
    "&kmAboveBelow=", str(ab),
    "&kmLeftRight=", str(lr) ]) )
````

**Iterate over groups of dates, request subsets from the REST API, and append to a list of responses:**

````python
chunks = list(chunk(modis_dates, 10))
subsets = []
for i, c in enumerate(chunks):     
    print("[ "+str(i+1)+" / "+str(len(chunks))+" ] "+c[0]+" - "+c[-1])
    _url = getSubsetURL(prod, lat, lng, band, c[0], c[-1], ab, lr)
    _response = requests.get(_url , headers = header)
    subsets.append( json.loads(_response.text) )
````

**Use dictionary comprehension to get some spatial metadata from the first subset in our list:**

````python
meta = { key:value for key,value in subsets[0].items() if key != "subset" }
meta
````

**Use the metadata to make lists of x and y coordinates:**          
`f(ncols,nrows): n * cellsize + corner_coordinate`

````python
xcoordinates = [float(meta['xllcorner'])]+[ i*meta['cellsize']+float(meta['xllcorner']) for i in range(1,meta['ncols']) ]
ycoordinates = [float(meta['yllcorner'])]+[ i*meta['cellsize']+float(meta['yllcorner']) for i in range(1,meta['nrows']) ]
print(str([xcoordinates[0],ycoordinates[0]]))
````

**Iterate over the list of subsets and collect the data in a dictionary:**

````python
data = {}
for i in subsets:                                     # iterate over list of subsets
    for j in i['subset']:                             # iterate over list of bands
        data[ j['calendar_date'] ] = np.array( j['data'] ).reshape( meta['nrows'], meta['ncols'] )
````

**Make a `gdal.Raster` object and add the `numpy` arrays to bands:**

GDAL data type codes:
```python
dtype = {
    "uint8" : 1 ,
    "int8":1 ,
    "uint16" : 2 ,
    "int16" : 3 ,
    "uint32" : 4 ,
    "int32" : 5 ,
    "float32" : 6 ,
    "float64" : 7 ,
    "complex64" : 10 ,
    "complex128" : 11
}
```

*Note: Remember that GDAL band numbering starts at 1 instead of 0!*

````python
# make a new raster object: 
ras = gdal.GetDriverByName( 'GTiff' ).Create( 
    'output-sinu.tif',                      # output filename
    meta['ncols'],                          # number of columns
    meta['nrows'],                          # number of rows
    len(data.keys()),                       # number of bands
    1                                       # data type (float32)
)

# set GDAL-style output geotransform (6-tuple)
ras.SetGeoTransform(( 
    xcoordinates[0],                        # x minimum
    meta['cellsize'],                       # x resolution
    0,                                      # x offset
    ycoordinates[-1],                       # y maximum
    0,                                      # y offset
    -meta['cellsize']                       # -y resolution
))

# make osr srs object using proj4 string; add wkt to raster
source_srs = osr.SpatialReference()
source_srs.ImportFromProj4( proj )
ras.SetProjection( source_srs.ExportToWkt() )

# loop over dict using enumerate() instead of .items; add data to each band
for index,key in enumerate(data):
    
    b = ras.GetRasterBand( index+1 ) # get raster band n
    b.WriteArray( data[key] )        # add data to band n
    b.SetDescription( key )          # set geotiff description as 
    b.SetNoDataValue( 255 )
        
ras.FlushCache()
ras
````

**Plot the first band of the new raster:**

````python
# use numpy's .ma.masked_greater to mask values outside valid range (>100)
masked_array = np.ma.masked_greater( np.flipud( data[list(data.keys())[0]] ), 100 ) * float( meta['scale'] )

fig, ax = plt.subplots(figsize=(10,8))
im = plt.pcolormesh(masked_array)
fig.colorbar(im)
plt.title('mean LAI [' + list(data.keys())[0] + ']')
````

Data stored in GDAL rasters must be extracted to `numpy` arrays for higher-level processing (hence all of the higher-level raster processing packages like rasterio). We can get the full raster stack as a 3-dimensional numpy array using `numpy.dstack()` and calculate the spatial mean over time using `numpy.mean()`.

**Plot a time series of mean LAI:**

````python
stacked_array = ras.ReadAsArray()              # get 3d numpy array
mean = np.mean(stacked_array, (1,2))           # get mean; remember Python numeration (0,1,2)
std = np.std(stacked_array, (1,2))             # get std
date = [ dt.datetime.strptime(d, "%Y-%m-%d").date() for d in list(data.keys()) ]

plt.rcParams['figure.figsize'] = (15,8)
fig, ax = plt.subplots()
ax.set_xlabel('Date')
ax.set_ylabel('mean LAI ['+meta['units']+']')
ax.plot(date, mean, c='black')
ax.fill_between(date, mean-std, mean+std, alpha=0.4)
fig.autofmt_xdate()
````

One of the more common reasons why you might want to put your subset into a GDAL raster is for easy warping.

**Use gdal.Warp to save an output GeoTIFF in geographic projection:**

````python
out = gdal.Warp( 'output-wgs84.tif', ras, dstSRS='EPSG:4326' )  # warp sinu raster object
ras, out = None, None                                           # close raster datasets
````
